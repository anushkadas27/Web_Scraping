```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
# Set up WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Run browser in headless mode (no UI)
driver = webdriver.Chrome(options=options)
pd.set_option('display.max_colwidth',None)
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
# Load the webpage (Replace 'your_local_html_file.html' with the actual file path or URL)
driver.get("https://anushkadas27.github.io/Job_Listing/")

# Extract full page source after JavaScript execution
html_content = driver.page_source

# Close the driver
driver.quit()
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://anushkadas27.github.io/Job_Listing/',headers=headers).text
```

```python
soup = BeautifulSoup(html_content, "lxml")
#soup=BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify()) 
```

```python
# Find all company info elements
company_elements = soup.find_all("div", class_="company-info")

# Extract company names
companies = [info.get_text().split("Company: ")[1].split(" | Location")[0] for info in company_elements]

# Print extracted company names
print("Extracted Company Names:")
for company in companies:
    print(company)
```

```python
# Find all company info elements
company_elements = soup.find_all("div", class_="company-info")

# Extract all company locations (including duplicates)
locations = []  # Using a list to keep all entries
for element in company_elements:
    text = element.get_text(strip=True)
    if "Location: " in text:
        location = text.split("Location: ")[1]  # Extract location
        locations.append(location)  # Append to list to keep duplicates

# Print all company locations
print("All Company Locations:")
for loc in locations:
    print(loc)
```

```python
# Find all job-title info elements
job_title_elements = soup.find_all("div", class_="job-title")

# Extract company names
job_titles = [info.get_text().split(" | Location")[0] for info in job_title_elements]

# Print extracted company names
print("Extracted Company Names:")
for job_title in job_titles:
    print(job_title)
```

```python
# Find all job detail elements
job_detail_elements = soup.find_all("div", class_="job-details")

# Extract all ratings (keeping duplicates)
ratings = []  # Using a list instead of a set
for element in job_detail_elements:
    text = element.get_text(strip=True)
    if "Rating: " in text:
        rating = text.split("Rating: ")[1].split(" /")[0]  # Extract rating value
        ratings.append(rating)  # Append to list to keep duplicates

# Print all ratings
print("All Ratings:")
for rate in ratings:
    print(rate)
```

```python
# Extract all salary values (keeping duplicates)
salaries = []  # Using a list instead of a set
for element in job_detail_elements:
    text = element.get_text(strip=True)
    if "Salary: " in text:
        salary = text.split("Salary: ")[1].split(" per year")[0]  # Extract salary value
        salaries.append(salary)  # Append to list to keep duplicates

# Print all salaries
print("All Salaries:")
for sal in salaries:
    print(sal)
```

```python
# Find all application stats elements
application_elements = soup.find_all("div", class_="application-stats")

# Extract all "Applied" values (keeping duplicates)
applied_counts = []  # Using a list instead of a set
for element in application_elements:
    text = element.get_text(strip=True)
    if "Applied: " in text:
        applied = text.split("Applied: ")[1].split(" |")[0]  # Extract applied value
        applied_counts.append(applied)  # Append to list to keep duplicates

# Print all applied values
print("All Applied Counts:")
for count in applied_counts:
    print(count)
```

```python
# Extract all "Interviews" values (keeping duplicates)
interviews_counts = []  # Using a list instead of a set
for element in application_elements:
    text = element.get_text(strip=True)
    if "Interviews: " in text:
        interviews = text.split("Interviews: ")[1].split(" |")[0]  # Extract interviews value
        interviews_counts.append(interviews)  # Append to list to keep duplicates

# Print all interviews values
print("All Interviews Counts:")
for count in interviews_counts:
    print(count)
```

```python
# Find all application stats elements
application_elements = soup.find_all("div", class_="application-stats")

# Extract all "Hired" values (keeping duplicates)
hired_counts = []  # Using a list instead of a set
for element in application_elements:
    text = element.get_text(strip=True)
    if "Hired: " in text:
        hired = text.split("Hired: ")[1]  # Extract hired value
        hired_counts.append(hired)  # Append to list to keep duplicates

# Print all hired values
print("All Hired Counts:")
for count in hired_counts:
    print(count)
```

```python
# Extracting  data  ensuring  proper alignment
job_titles = [job.get_text(strip=True) for job in soup.find_all("div", class_="job-title")]
companies = [info.get_text(strip=True).split("|")[0].split("Company: ")[1].strip() for info in soup.find_all("div", class_="company-info")]
locations = [info.get_text(strip=True).split("| Location: ")[1].strip() for info in soup.find_all("div", class_="company-info")]
ratings = [details.get_text(strip=True).split("Rating: ")[1].split(" /")[0] for details in soup.find_all("div", class_="job-details")]
salaries = [details.get_text(strip=True).split("Salary: ")[1].split(" per year")[0].strip() for details in soup.find_all("div", class_="job-details")]
applied_counts = [stats.get_text(strip=True).split("Applied: ")[1].split(" |")[0] for stats in soup.find_all("div", class_="application-stats")]
interviews_counts = [stats.get_text(strip=True).split("Interviews: ")[1].split(" |")[0] for stats in soup.find_all("div", class_="application-stats")]
hired_counts = [stats.get_text(strip=True).split("Hired: ")[1] for stats in soup.find_all("div", class_="application-stats")]

# Ensure all lists have equal length
max_length = max(len(job_titles), len(companies), len(locations), len(ratings), len(salaries), len(applied_counts), len(interviews_counts), len(hired_counts))

# Fill missing values with None
job_titles += [None] * (max_length - len(job_titles))
companies += [None] * (max_length - len(companies))
locations += [None] * (max_length - len(locations))
ratings += [None] * (max_length - len(ratings))
salaries += [None] * (max_length - len(salaries))
applied_counts += [None] * (max_length - len(applied_counts))
interviews_counts += [None] * (max_length - len(interviews_counts))
hired_counts += [None] * (max_length - len(hired_counts))

# Create the DataFrame
df = pd.DataFrame({
    "Job Title": job_titles,
    "Company": companies,
    "Location": locations,
    "Rating": ratings,
    "Salary": salaries,
    "Applied": applied_counts,
    "Interviews": interviews_counts,
    "Hired": hired_counts
})

# Display DataFrame
print(df)
```

```python
df.to_csv('Company_Details.csv',index=False)
```
